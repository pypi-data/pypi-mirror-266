```python
import pandas
from pandas_datareader import data as pdr
import yfinance as yfin
import datetime as dt
```

```python
yfin.pdr_override()

djia = pdr.get_data_yahoo('DJI', start='1986-01-01', end='2023-07-31')
```

```python
djia['CloseDiff'] = djia['Close'].diff(1)
```

```python
djia['CloseDiffPerc'] = djia['CloseDiff'] / djia['Close'].shift(-1) * 100
```

```python
djia['NoRecentCloseBelowX'] = djia['CloseDiffPerc'].transform(lambda x: x.rolling(44).min() > -1)
djia[djia['NoRecentCloseBelowX'] == True]
```
