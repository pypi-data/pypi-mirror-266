```python
import pandas as pd
manual_data = pd.read_csv("../manual_data/indices.csv")
data = NseIndicesNewReader().read(ForDateCriteria(date(2024, 3, 5)))
```

```python
old_ids = manual_data['Identifier'].unique()
new_ids = data['Identifier'].unique()
mapping = {}
for old_id in old_ids:
  matching_new_ids = list(filter(lambda x: x.upper() == old_id.upper(), new_ids))
  if len(matching_new_ids):
    mapping[old_id] = matching_new_ids[0]
```
